2. Determine the percentage of both male and female in each group

```python
df_per = df.groupby('gender')['race/ethnicity'].value_counts(normalize=True)
df_per = df_per.mul(100).rename('Percent').reset_index()
```

```python
df = sns.catplot(x='gender', y='Percent', hue='race/ethnicity', kind='bar', data=df_per)
df.ax.set_ylim(0,100)

for p in df.ax.patches:
    per = str(p.get_height().round(1)) + '%'
    per_x = p.get_x()
    per_y = p.get_height()
    df.ax.text(per_x,per_y,per)
```
